```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [
    u'Pattern matching (with tree database)',
    u'Find the star pattern',
    u'Choose Reference Star',
    u'Centroiding']
y = [20,30,54,17688]

fig, ax = plt.subplots()    
width = 0.75 # the width of the bars 
ind = np.arange(len(y))  # the x locations for the groups
ax.barh(ind, y, width, color="blue")
ax.set_yticks(ind+width/2)
ax.set_yticklabels(x, minor=False)

plt.title('Module Profiling Result')
plt.xlabel('Runtime(nanoseconds)')
plt.ylabel('Modules')

for i, v in enumerate(y):
    ax.text(v + 3, i + .25, str(v), color='blue', fontweight='bold')
    
plt.show()
```

```python
x = [
    u'Pattern Matching',
    u'Finding Star Pattern',
    u'Choosing Reference Star',
    u'Centroiding'
    ]

a = [89,30,54,17688]
b = [155,109,105,18061]

fig, ax = plt.subplots()    
width = 0.2 # the width of the bars 
ind = np.arange(len(a))  # the x locations for the groups

ax.barh(ind+0.2, a, width, color="blue", label='Average')
ax.barh(ind, b, width, color="red", label='Worstcase')


ax.set_yticks(ind+width/2)
ax.set_yticklabels(x, minor=False)

plt.title('Software runtime')
plt.legend(bbox_to_anchor=(0.95, 0.2))
plt.xlabel('Runtime (microseconds)')
plt.ylabel('Modules')

for i, v in enumerate(a):
    ax.text(v + 0.2, i+0.15, str(v), color='blue', fontweight='bold')
for i, v in enumerate(b):
    ax.text(v + 0.2, i-0.05, str(v), color='red', fontweight='bold')

plt.show()
```
